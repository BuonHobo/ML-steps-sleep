Noto che la remduration manca praticamente sempre, mentre il resto dei dati sul sonno sono presenti sempre tutti insieme. Trovo un modo per calcolare il punteggio che faccia uso di questi dati.

```python
import pandas as pd
```

```python
steps = pd.read_csv("../data/dati.csv.gz")[
    ["uuid", "date", "steps", 'sleepduration', 'lightduration',
     'deepduration']]  # Il dataset qui è abbastanza denso, per ora può andare bene eliminare le righe con valori mancanti
```

```python
def sleep_score(x):
    optimal_light_ratio = .5
    optimal_deep_ratio = .25

    # Penalità sul light sleep X Penalità sul deep sleep X Penalità sulla durata del sonno
    x["sleep_score"] = (1 - abs(x["lightduration"] / x["sleepduration"] - optimal_light_ratio)) * (
            1 - abs(x["deepduration"] / x["sleepduration"] - optimal_deep_ratio)) * (
                               1 - abs(x["sleepduration"] / x["sleepduration"].mean() - 1))
    return x


steps = steps.groupby("uuid").apply(sleep_score, include_groups=False)
steps.sample(100)
```

```python
steps = steps.rename_axis(["uuid", "id"]).reset_index().drop(
    columns=["id"])  # Per un csv più pulito, infatti il groupby aggiunge una colonna che non voglio
steps.head()
```
